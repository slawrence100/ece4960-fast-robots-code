## Kalman Filter

### Setup

```python
#We assume uncorrelated noise, therefore a diagonal matrix works.
sampling_rate = len(tof_times) / 2.5
sigma_1 = np.sqrt(0.1**2 * sampling_rate) # Trust in modeled position
sigma_2 = np.sqrt(0.01**2 * sampling_rate) # Trust in modeled speed
sigma_4 = 0.025 # measurement noise
sig_u=np.array([[sigma_1**2,0],[0,sigma_2**2]]) # u = dist from wall
sig_z=np.array([[sigma_4**2]]) # measurement noise
print(f"sampling_rate={sampling_rate}")
print(f"sigma_1={sigma_1} m, sigma_2={sigma_2} m/s, sigma_4={sigma_4} m")

d = 1 / abs(steady_state_vel)
m = -d * rise_time / np.log(0.1)

A = np.array([[0,1], [0, -1*d/m]])
B = np.array([[0], [1/m]])
C = np.array([[-1,0]])

print(f"d={d}, m={m}, rise time={rise_time} s")
print(f"Sigma_u:{sig_u}")
print(f"Sigma_z:{sig_z}")
print(f"A: {A}")
print(f"B: {B}")
print(f"C: {C}")

x = np.array([[tof_vals[1]],[0]])
delta_t = 1 / sampling_rate
A_d = np.eye(2) + delta_t * A
B_d = delta_t * B
print(f"delta_t: {delta_t}")
print(f"A_d: {A_d}")
print(f"B_d: {B_d}")
"""
mu: Best guess of where we are so far
sigma: Uncertainty so far
u: motor power input
y: sensor reading
"""
def kf(mu_prev,sigma_prev,u,y):
    
    mu_p = A.dot(mu_prev) + B.dot(u) 
    sigma_p = A.dot(sigma_prev.dot(A.transpose())) + sig_u
    
    y_m = y-C.dot(mu_p)
    sigma_m = C.dot(sigma_p.dot(C.transpose())) + sig_z
    kkf_gain = sigma_p.dot(C.transpose().dot(np.linalg.inv(sigma_m)))

    mu = mu_p + kkf_gain.dot(y_m)    
    sigma=(np.eye(2)-kkf_gain.dot(C)).dot(sigma_p)

    return mu,sigma
```

### Execute on Test Data

```python
# Set initial mu and sigma
mu_0 = x
sigma_0 = np.array([[sigma_1,0],[0,0]])
print(f"mu_0: {mu_0}")
print(f"sigma_0: {sigma_0}")

mu = [mu_0]
sigma = [sigma_0]

for u,y in zip(motor_vals[1:],tof_vals[1:]):
  mu_new, sigma_new = kf(mu[-1], sigma[-1], u/max(motor_vals), y)
  mu.append(mu_new)
  sigma.append(sigma_new)

kf_dist = []
kf_vel = []
for m in mu:
  kf_dist.append(abs(m[0][0]))
  kf_vel.append(abs(m[1][0]))
```
